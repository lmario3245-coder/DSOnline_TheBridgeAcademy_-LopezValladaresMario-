```python
# ==========================
# PARTE I: Nuevos alumnos y algoritmo
# ==========================
import numpy as np

# 1️⃣ Creamos array original de alumnos
alumnos_aidone = np.array([
    [5, 4],
    [2, -3],
    [3, 5],
    [-5, 4],
    [0, -1],
    [3, 2],
    [-1, 1],
    [5, -2],
    [-5, 3],
    [-5, 1]
])

# Nuevos alumnos: Teresa y Udo
teresa = np.array([[-2, -5]])
udo = np.array([[0, 4]])

# Concatenamos los nuevos alumnos
alumnos_aidone = np.vstack([alumnos_aidone, teresa, udo])
print("Array de alumnos completo:\n", alumnos_aidone)

# Referencias y lista de alumnos
alumnos_ref = ["Rodrigo","Gabriela","Mateo"]
lista_alumnos = ['Rodrigo', 'Lucia', 'Alejandro', 'Valeria', 'Javier', 'Camila', 'Diego', 
                 'Gabriela', 'Mateo', 'Sofía','Teresa','Udo']
curso_asignado = ["CAI","ER","NMPC"]

# Inicializamos la matriz de reparto (12 alumnos x 3 cursos)
matriz_reparto = np.zeros((12,3))

# Función de distancia Euclidiana
def distancia_2d(v1, v2):
    return np.linalg.norm(v1 - v2)

# Asignación de alumnos a cursos
for i_vector, alumno in enumerate(lista_alumnos):
    vector_alumno = alumnos_aidone[i_vector]
    
    # Distancias a los alumnos de referencia
    vector_distancias = np.array([distancia_2d(vector_alumno, alumnos_aidone[lista_alumnos.index(ref)]) 
                                  for ref in alumnos_ref])
    
    # Elegimos curso con menor distancia
    indice_curso_elegido = np.argmin(vector_distancias)
    matriz_reparto[i_vector, indice_curso_elegido] = 1
    
    print(f"{alumno} será asignado a {curso_asignado[indice_curso_elegido]}")

print("\nMatriz de reparto final:\n", matriz_reparto)
```

```python
# ==========================
# Crear tablero 10x10
# ==========================
tablero = np.full((10,10), " ")  # lleno de espacios
print(tablero)
# Posicionar barcos iniciales
barco1 = [(0,1), (1,1)]              # Barco de 2 posiciones
barco2 = [(1,3), (1,4), (1,5), (1,6)] # Barco de 4 posiciones

for x,y in barco1 + barco2:
    tablero[x,y] = "O"

print("Tablero con barcos iniciales:\n", tablero)
# Función para disparos
def disparo(tablero, fila, col):
    if tablero[fila, col] == "O":
        tablero[fila, col] = "X"
    elif tablero[fila, col] == " ":
        tablero[fila, col] = "-"
disparo(tablero, 0, 1)  # acierto
disparo(tablero, 0, 0)  # agua

print("Tablero después de disparos:\n", tablero)
# Función para colocar barcos aleatoriamente
import random

def colocar_barco_aleatorio(tablero, longitud):
    filas, cols = tablero.shape
    colocado = False
    
    while not colocado:
        fila = random.randint(0, filas-1)
        col = random.randint(0, cols-1)
        orientacion = random.choice(["N","S","E","O"])
        
        posiciones = []
        for i in range(longitud):
            if orientacion == "N":
                x, y = fila-i, col
            elif orientacion == "S":
                x, y = fila+i, col
            elif orientacion == "E":
                x, y = fila, col+i
            else:  # "O"
                x, y = fila, col-i
            
            if x<0 or x>=filas or y<0 or y>=cols or tablero[x,y] != " ":
                break
            posiciones.append((x,y))
        
        if len(posiciones) == longitud:
            for x,y in posiciones:
                tablero[x,y] = "O"
            colocado = True
colocar_barco_aleatorio(tablero, 4)
print("Tablero después de barco aleatorio:\n", tablero)
```
