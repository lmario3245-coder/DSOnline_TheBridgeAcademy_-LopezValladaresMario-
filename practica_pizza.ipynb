```python
# --- PARTE I: Chatbot de pizzas ---

# Saludo inicial del chatbot
print("Buenas tardes, bienvenido al servicio de pedido online, el precio de nuestras pizzas es de 8,95 para la familiar o de 9,90 para la mediana.")
print("¿Cuántas pizzas familiares desea?")

# Usuario introduce número de pizzas familiares
pizz_familiar = int(input())

# Pregunta siguiente
print("Estupendo, ¿cuántas pizzas medianas desea?")
pizz_mediana = int(input())

# Confirmación del pedido
print(f"Estupendo, se están preparando {pizz_familiar} pizzas familiares y {pizz_mediana} pizzas medianas. Dígame su dirección:")

# Usuario introduce dirección
direcc = input()

# Precios de las pizzas
precio_familiar = 8.95
precio_mediana = 9.90

# Cálculo del total
total = pizz_familiar * precio_familiar + pizz_mediana * precio_mediana
total = round(total, 2)  # Redondeamos a 2 decimales

# Número total de pizzas
num_total = pizz_familiar + pizz_mediana

# Mensaje final del chatbot
print(f"Le mandaremos las {num_total} pizzas a la dirección '{direcc}'. Serán {total} €. Muchas gracias por su pedido.")
```

```python
# --- PARTE II: Números enteros ---

# Solicitar 3 números enteros separados por comas
entrada = input("Introduce 3 números enteros separados por comas (ej: 1,2,3): ")

# Convertir la entrada en una lista de enteros
numeros = [int(x) for x in entrada.split(",")]

# Determinar máximo, mínimo y suma total
num_max = max(numeros)
num_min = min(numeros)
suma_total = sum(numeros)

# Comprobar la afirmación
afirmacion = (num_max - num_min) == (suma_total / 5)

# Mostrar resultados
print(f"Números introducidos: {numeros}")
print(f"Número máximo: {num_max}")
print(f"Número mínimo: {num_min}")
print(f"Suma total: {suma_total}")
print(f"¿Se cumple la afirmación 'máximo - mínimo = suma/5'?: {afirmacion}")
```
