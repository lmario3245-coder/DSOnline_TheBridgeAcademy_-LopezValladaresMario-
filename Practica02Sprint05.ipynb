```python
import os

# Listar archivos en la carpeta data
for fichero in os.listdir("./data"):
    print(fichero)

# Asignar rutas a variables según los ficheros encontrados
fichero_texto = "./data/nombre_fichero.txt"   # remplazar con el nombre real
fichero_excel = "./data/nombre_fichero.xlsx"  # remplazar con el nombre real
fichero_xml = "./data/nombre_fichero.xml"     # remplazar con el nombre real
```

```python
import pandas as pd

# Leer Excel en un DataFrame
df_excel = pd.read_excel(fichero_excel)

# Renombrar columnas
df_excel.rename(columns={df_excel.columns[0]:"revenue", "num":"sales"}, inplace=True)

# Mostrar info general
print(df_excel.info())
print(df_excel.describe())

# Convertir la columna "date" a datetime y ponerla como índice
df_excel["date"] = pd.to_datetime(df_excel["date"])
df_excel.set_index("date", inplace=True)

# Ventas y revenue agrupadas por ciudad entre 2017 y 2018
df_filtrado = df_excel[(df_excel.index.year >= 2017) & (df_excel.index.year <= 2018)]
ventas_por_ciudad = df_filtrado.groupby("city")["sales"].sum()
revenue_por_ciudad = df_filtrado.groupby("city")["revenue"].sum()

print(ventas_por_ciudad)
print(revenue_por_ciudad)

# Sobreescribir el fichero Excel con la hoja "clean_data" y sin índice
df_excel.reset_index().to_excel(fichero_excel, sheet_name="clean_data", index=False)
# Crear diccionario vacío
spam_dict = {"from":[], "to":[], "is_spam":[], "confidence":[]}

# Leer fichero y procesar
with open(fichero_texto, "r") as f:
    lines = f.readlines()

# Mostrar primeras 60 líneas
print("".join(lines[:60]))

# Recorrer líneas y rellenar el diccionario
for line in lines:
    line = line.strip()
    if line.startswith("From:"):
        spam_dict["from"].append(line.split("From:")[1].strip())
    elif line.startswith("To:"):
        spam_dict["to"].append(line.split("To:")[1].strip())
    elif line.startswith("X-DSPAM-Result:"):
        spam_dict["is_spam"].append(line.split("X-DSPAM-Result:")[1].strip())
    elif line.startswith("X-DSPAM-Confidence:"):
        spam_dict["confidence"].append(float(line.split("X-DSPAM-Confidence:")[1].strip()))

# Comprobar longitud de las listas
for key, value in spam_dict.items():
    print(key, len(value))

# Crear DataFrame
df_spam = pd.DataFrame(spam_dict)

# Mostrar correos marcados como spam y confianza media
print(df_spam[df_spam["is_spam"]=="spam"])
print("Confianza media:", df_spam["confidence"].mean())

# Guardar en CSV con "|" como separador
df_spam.to_csv("./data/spam.csv", sep="|", index=False)
import xml.etree.ElementTree as ET
import pandas as pd

# Leer XML
arbol = ET.parse(fichero_xml)
ejer_raiz = arbol.getroot()

# Mostrar etiqueta y atributos del nodo raíz
print("Etiqueta raíz:", ejer_raiz.tag)
print("Atributos raíz:", ejer_raiz.attrib)

# Número de hijos y nietos
print("Número de hijos:", len(ejer_raiz))
for hijo in ejer_raiz:
    print("Etiqueta hijo:", hijo.tag, "Número de nietos:", len(hijo))

# Encontrar un nodo 'Event' y mostrar atributos
primer_evento = ejer_raiz.find("Event")
print(primer_evento.attrib)

# Crear DataFrame con todos los eventos
eventos_data = []
for evento in ejer_raiz.findall("Event"):
    eventos_data.append(evento.attrib)

df_eventos = pd.DataFrame(eventos_data)
print(df_eventos.head())
```
