```python
import sqlite3
import pandas as pd

# Conexión a la base de datos
conn = sqlite3.connect("./data/chinook.db")
cursor = conn.cursor()

# Función para ejecutar queries y devolver DataFrame
def query_to_df(sql):
    return pd.read_sql(sql, conn)
```

```python
SELECT CustomerId, FirstName || ' ' || LastName AS Nombre, Country
FROM customers
WHERE Country IN ('Canada','Brazil');
df_customers = pd.read_sql("SELECT * FROM customers", conn)
df_customers[(df_customers['Country'].isin(['Canada','Brazil']))][['CustomerId','FirstName','LastName','Country']]
```

```python
SELECT DISTINCT BillingCountry
FROM invoices;
df_invoices = pd.read_sql("SELECT * FROM invoices", conn)
df_invoices['BillingCountry'].drop_duplicates()
SELECT State, COUNT(*) 
FROM customers
WHERE Country='USA'
GROUP BY State
ORDER BY 2 DESC;
df_customers_usa = df_customers[df_customers['Country']=='USA']
df_customers_usa.groupby('State').size().sort_values(ascending=False)
```

```python
SELECT t.Name AS Cancion, g.Name AS Genero
FROM tracks t
LEFT JOIN genres g ON t.GenreId = g.GenreId;
df_tracks = pd.read_sql("SELECT * FROM tracks", conn)
df_genres = pd.read_sql("SELECT * FROM genres", conn)
df_join = df_tracks.merge(df_genres, left_on="GenreId", right_on="GenreId", how="left")
df_join.head()
df_join = df_join.rename(columns={"Name_x":"Cancion","Name_y":"Genero"})
df_join = df_join[['Cancion','Genero']]
```

```python
SELECT g.Name AS Genero, COUNT(*) 
FROM tracks t
LEFT JOIN genres g ON t.GenreId = g.GenreId
GROUP BY g.Name
ORDER BY 2 DESC;
df_join.groupby('Genero').size().sort_values(ascending=False)
SELECT c.FirstName || ' ' || c.LastName AS Nombre, c.City, c.Country, i.InvoiceId, i.InvoiceDate
FROM customers c
INNER JOIN invoices i ON c.CustomerId = i.CustomerId
WHERE c.Country='Germany';
df_customers = pd.read_sql("SELECT * FROM customers", conn)
df_invoices = pd.read_sql("SELECT * FROM invoices", conn)

df_merge = df_customers.merge(df_invoices, left_on="CustomerId", right_on="CustomerId", how="inner")
df_merge[df_merge['Country']=='Germany'][['FirstName','LastName','City','Country','InvoiceId','InvoiceDate']]
```

```python
SELECT c.FirstName || ' ' || c.LastName AS Nombre, COUNT(i.InvoiceId) AS NumCompras
FROM customers c
INNER JOIN invoices i ON c.CustomerId = i.CustomerId
WHERE c.Country='Brazil'
GROUP BY c.CustomerId;
df_brasil = df_customers.merge(df_invoices, on="CustomerId", how="inner")
df_brasil[df_brasil['Country']=='Brazil'].groupby(['CustomerId','FirstName','LastName']).size()
```
